# Campo Eléctrico generado en $\vec{r}$ por carga puntual $q^{\prime}$ ubicada en $\vec{r}^{\prime}$

$$\vec{E}(\vec{r})=k_{e}q^{\prime}\frac{(\vec{r}-\vec{r}^{\prime})}{||\vec{r}-\vec{r}^{\prime}||^3}.$$

# Campo eléctrico de carga puntual $q^{\prime}$ ubicada en origen de sistema de referencia donde $\vec{r}^{\prime}=\vec{0}$

$$\vec{E}(\vec{r})=k_{e}q^{\prime}\frac{\vec{r}}{||\vec{r}||^3}.$$

```python
#Escrito por Nataly Ibarra en marzo de 2017
#Ante cualquier duda o comentario contactar
#a natalynicole.ibarravera@gmail.com
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def Ex(x,y):
    return kq*x/(x**2+y**2)**(3./2)
def Ey(x,y):
    return kq*y/(x**2+y**2)**(3./2)
#kq_p = 1
#kq_m = -1
```

```python
n = 48
x = np.linspace(-np.pi,np.pi,n)
y = np.linspace(-np.pi,np.pi,n)
X, Y = np.meshgrid(x,y)
```

```python
kq = 1.
fig = plt.figure(figsize=(6,6))
Q = plt.quiver(X, Y, Ex(X,Y), Ey(X,Y), units='width')
plt.xlim(-np.pi,np.pi)
plt.ylim(-np.pi,np.pi)
plt.xlabel('$x$',fontsize=18)
plt.ylabel('$y$',fontsize=18)
plt.title(r'$\vec{E}(\vec{r})$ con $q>0$',fontsize=18)
c1 = plt.Circle((0, 0), 0.15, color='red')
plt.gcf().gca().add_artist(c1)
```

```python
kq = -1.
fig = plt.figure(figsize=(6,6))
Q = plt.quiver(X, Y, Ex(X,Y), Ey(X,Y), units='width')
plt.xlim(-np.pi,np.pi)
plt.ylim(-np.pi,np.pi)
plt.xlabel('$x$',fontsize=18)
plt.ylabel('$y$',fontsize=18)
plt.title(r'$\vec{E}(\vec{r})$ con $q<0$',fontsize=18)
c2 = plt.Circle((0, 0), 0.15, color='blue')
plt.gcf().gca().add_artist(c2)
```

En coordenadas esféricas $(r,\theta,\varphi)$ en torno a la carga $q$: 
$$\vec{E}(\vec{r})=\frac{k_{e} q}{r^2}\hat{r}$$

Definiendo la componente radial
$$E_{r}(r):=\frac{k_{e} q}{r^2},$$
entonces
$$\vec{E}(\vec{r})=E(r)\hat{r}.$$

```python
def Er(r):
    return kq/r**2
r = np.linspace(0.1,2,100)
```

```python
plt.figure(figsize=(6,6))
kq = 1.
plt.plot(r,Er(r),label=r'$\frac{k_{e}q_{+}}{r^2}$')
kq = -1.
plt.plot(r,Er(r),label=r'$\frac{k_{e}q_{-}}{r^2}$')
plt.grid()
plt.xlabel(r'$r$',fontsize=18)
plt.ylabel(r'$E_{r}(r)$',fontsize=18)
plt.legend(fontsize=18)
```
