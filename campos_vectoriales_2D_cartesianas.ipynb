# Gradiente de Campo Escalar

En coordenadas cartesianas el gradiente de un campo escalar $f:=f(x,y)$ se calcula como:
$$ \vec{\nabla}f=\frac{\partial f}{\partial x}\,\hat{x}+\frac{\partial f}{\partial y}\,\hat{y}.$$

En este notebook se han graficado una serie de campos vectoriales en el plano 2D. 

Se ha elegido para graficar un dominio $x, y\in [-\pi,\pi]$.

```python
n = 35
x = np.linspace(-np.pi,np.pi,n)
y = np.linspace(-np.pi,np.pi,n)
X, Y = np.meshgrid(x,y)
```

# $\vec{\nabla} f_1(x,y)=1\,\hat{x}+0\,\hat{y}$

```python
Cx = X/X
Cy = 0*Y
fig = plt.figure(figsize=(6,6))
Q = plt.quiver(X, Y, Cx, Cy, units='width')
plt.xlim(-np.pi,np.pi)
plt.ylim(-np.pi,np.pi)
plt.xlabel('$x$',fontsize=18)
plt.ylabel('$y$',fontsize=18)
plt.title(r'$\vec{\nabla}f_1(x,y)=1\,\hat{x}+0\,\hat{y}$',fontsize=18)
```

```python
Cx = 0*X
Cy = Y/Y
fig = plt.figure(figsize=(6,6))
Q = plt.quiver(X, Y, Cx, Cy, units='width')
plt.xlim(-np.pi,np.pi)
plt.ylim(-np.pi,np.pi)
plt.xlabel('$x$',fontsize=18)
plt.ylabel('$y$',fontsize=18)
plt.title(r'$\vec{\nabla}f_2(x,y)=0\,\hat{x}+1\,\hat{y}$',fontsize=18)
```

```python
Cx = Y
Cy = X
fig = plt.figure(figsize=(6,6))
Q = plt.quiver(X, Y, Cx, Cy, units='width')
plt.xlim(-np.pi,np.pi)
plt.ylim(-np.pi,np.pi)
plt.xlabel('$x$',fontsize=18)
plt.ylabel('$y$',fontsize=18)
plt.title(r'$\vec{\nabla}f_3(x,y)=y\,\hat{x}+x\,\hat{y}$',fontsize=18)
```

```python
Cx = 2*X
Cy = 0*Y
fig = plt.figure(figsize=(6,6))
Q = plt.quiver(X, Y, Cx, Cy, units='width')
plt.xlim(-np.pi,np.pi)
plt.ylim(-np.pi,np.pi)
plt.xlabel('$x$',fontsize=18)
plt.ylabel('$y$',fontsize=18)
plt.title(r'$\vec{\nabla}f_4(x,y)=2x\,\hat{x}+0\,\hat{y}$',fontsize=18)
```

```python
Cx = 0*X
Cy = 2*Y
fig = plt.figure(figsize=(6,6))
Q = plt.quiver(X, Y, Cx, Cy, units='width')
plt.xlim(-np.pi,np.pi)
plt.ylim(-np.pi,np.pi)
plt.xlabel('$x$',fontsize=18)
plt.ylabel('$y$',fontsize=18)
plt.title(r'$\vec{\nabla}f_5(x,y)=0\,\hat{x}+2y\,\hat{y}$',fontsize=18)
```

```python
Cx = np.cos(X)
Cy = 0*Y
fig = plt.figure(figsize=(6,6))
Q = plt.quiver(X, Y, Cx, Cy, units='width')
plt.xlim(-np.pi,np.pi)
plt.ylim(-np.pi,np.pi)
plt.xlabel('$x$',fontsize=18)
plt.ylabel('$y$',fontsize=18)
plt.title(r'$\vec{\nabla}f_7(x,y)=\cos(x)\,\hat{x}+0\,\hat{y}$',fontsize=18)
```

```python
Cx = Y*np.cos(X*Y)
Cy = X*np.cos(X*Y)
fig = plt.figure(figsize=(6,6))
Q = plt.quiver(X, Y, Cx, Cy, units='width')
plt.xlim(-np.pi,np.pi)
plt.ylim(-np.pi,np.pi)
plt.xlabel('$x$',fontsize=18)
plt.ylabel('$y$',fontsize=18)
plt.title(r'$\vec{\nabla}f_8(x,y)=y\,\cos(xy)\,\hat{x}+x\,\cos(xy)\,\hat{y}$',fontsize=18)
```

```python
Cx = 2*X*np.cos(X**2+Y**2)
Cy = 2*Y*np.cos(X**2+Y**2)
fig = plt.figure(figsize=(6,6))
Q = plt.quiver(X, Y, Cx, Cy, units='width')
plt.xlim(-np.pi,np.pi)
plt.ylim(-np.pi,np.pi)
plt.xlabel('$x$',fontsize=18)
plt.ylabel('$y$',fontsize=18)
plt.title(r'$\vec{\nabla}f_9(x,y)=2x\,\cos(x^2+y^2)\,\hat{x}+2y\,\cos(x^2+y^2)\,\hat{y}$',fontsize=18)
```
