# Campo eléctrico de Dipolo eléctrico

Configuración: Dos cargas $q_{1}=q$ y $q_{2}=-q$ ubicadas en las posiciónes $\vec{r}_{1}=(d/2)\,\hat{y}$ y $\vec{r}_{2}=-(d/2)\,\hat{y}$ respectivamente.

En la posición $\vec{r}=x\hat{x}+y\hat{y}$, el campo eléctrico del dipolo se encuentra dado por

$$\vec{E}(\vec{r})=k_{e}q_{1}\frac{(\vec{r}-\vec{r}_{1})}{||\vec{r}-\vec{r}_{1}||^3}+k_{e}q_{2}\frac{(\vec{r}-\vec{r}_{2})}{||\vec{r}-\vec{r}_{2}||^3},$$

$$ \vec{E}(\vec{r})=k_{e}q\left(\frac{x\hat{x}+(y-d/2)\hat{y}}{(x^2+(y-d/2)^2)^{3/2}}-\frac{x\hat{x}+(y+d/2)\hat{y}}{(x^2+(y+d/2)^2)^{3/2}}\right).$$

```python
#Escrito por Nataly Ibarra en marzo de 2017
#Ante cualquier duda o comentario contactar
#a natalynicole.ibarravera@gmail.com
#Modificado en agosto de 2018
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.matplotlib.rcParams.update({'font.size': 18, 'font.family': 'serif'})
```

```python
def Ex(x,y):
    return kq*(x/(x**2.+(y-d/2.)**2.)**(3./2.)-x/(x**2.+(y+d/2.)**2.)**(3./2.))
def Ey(x,y):
    return kq*((y-d/2.)/(x**2.+(y-d/2.)**2.)**(3./2.)-(y+d/2.)/(x**2.+(y+d/2.)**2.)**(3./2.))
```

```python
kq = 1.
d = 3.
Ex(0,0),Ey(0,0)
```

```python
n = 48
x = np.linspace(-np.pi,np.pi,n)
y = np.linspace(-np.pi,np.pi,n)
X, Y = np.meshgrid(x,y)
```

```python
fig = plt.figure(figsize=(6,6))
Q = plt.quiver(X, Y, Ex(X,Y), Ey(X,Y), units='width')
plt.xlim(-np.pi,np.pi)
plt.ylim(-np.pi,np.pi)
plt.xlabel('$x$',fontsize=18)
plt.ylabel('$y$',fontsize=18)
plt.title(ur'Dipolo eléctrico',fontsize=18)
c1 = plt.Circle((0, d/2.), 0.15, color='red')
c2 = plt.Circle((0, -d/2.), 0.15, color='blue')
plt.gcf().gca().add_artist(c1)
plt.gcf().gca().add_artist(c2)
```

# Líneas de Campo eléctrico

```python
fig = plt.figure(figsize=(6,6))
Q = plt.streamplot(X, Y, Ex(X,Y), Ey(X,Y))
plt.xlim(-np.pi,np.pi)
plt.ylim(-np.pi,np.pi)
plt.xlabel('$x$',fontsize=18)
plt.ylabel('$y$',fontsize=18)
plt.title(ur'Dipolo eléctrico',fontsize=18)
c1 = plt.Circle((0, d/2.), 0.15, color='red')
c2 = plt.Circle((0, -d/2.), 0.15, color='blue')
plt.gcf().gca().add_artist(c1)
plt.gcf().gca().add_artist(c2)
```

# Si $\vec{r}$ está sobre el eje $x$:

$$\vec{E}(\vec{r})=-\frac{k_{e}qd}{(x^2+(d/2)^2)^{3/2}} \hat{y}.$$

```python
def Ey_x(x):
    return -(kq*d)/((x**2+(d/2)**2)**1.5)
```

```python
x0 = 2.0
print 'La componente $y$ del campo eléctrico en $\vec{r}=%2.2f \hat{x}$~m está dado por: %2.2f N/C'%(x0,Ey_x(x0))
```

```python
kq = 1.
d = 3.
x = np.linspace(-4,4,100)
plt.figure(figsize=(6,6))
plt.plot(x,Ey_x(x),lw=3)
plt.xlabel('$x$',fontsize=18)
plt.ylabel('$E_{y}(x)$',fontsize=18)
plt.grid()
```

# Si $\vec{r}$ está sobre el eje $y$:

$$\vec{E}(\vec{r})=k_{e}q \left( \frac{(y-d/2)}{((y-d/2)^2)^{3/2}} -\frac{(y+d/2)}{((y+d/2)^2)^{3/2}} \right)\hat{y}$$

Ahora bien,

$$\vec{E}(\vec{r})=k_{e}q \left( \frac{(y-d/2)}{((y-d/2)^2)^{3/2}} -\frac{(y+d/2)}{((y+d/2)^2)^{3/2}} \right)\hat{y}
=\left\{ 
\begin{eqnarray}
+k_{e}q \left( \frac{2yd}{(y-d/2)^2 (y+d/2)^2} \right)\hat{y},\qquad &\text{si } y>d/2.\\
-2k_{e}q \left( \frac{y^2+(d/2)^2}{(y-d/2)^2 (y+d/2)^2} \right)\hat{y},\qquad &\text{si } -d/2<y< d/2.\\
-k_{e}q \left( \frac{2yd}{(y-d/2)^2 (y+d/2)^2} \right)\hat{y},\qquad &\text{si } y< -d/2.
\end{eqnarray}
\right .$$

```python
def Eyy(y):
    return kq*((y-d/2.)/(((y-d/2.)**2)**(3./2.))-(y+d/2.)/(((y+d/2.)**2)**(3./2.)))
#
def Ey_y1(y):
    return 2.*y*(kq*d)/(((y-d/2.)**2.)*((y+d/2.)**2.))
def Ey_y2(y):
    return -2.*(kq)*(y**2+d**2/4.)/(((y-d/2.)**2.)*((y+d/2.)**2.))
def Ey_y3(y):
    return -2.*y*(kq*d)/(((y-d/2.)**2.)*((y+d/2.)**2.))
Eyyv = np.vectorize(Eyy)
Ey_yv1 = np.vectorize(Ey_y1)
Ey_yv2 = np.vectorize(Ey_y2)
Ey_yv3 = np.vectorize(Ey_y3)
```

```python
y = np.linspace(-4,4,101)
```

```python
#y = np.linspace(-4,4,1000)
plt.figure(figsize=(8,8))
plt.plot(y,Eyyv(y),'--',lw=3,label='$E_{y}(y)$',color='black')
plt.plot(y,Ey_yv1(y),'.-',label='$y>d/2$')
plt.plot(y,Ey_yv2(y),'.-',label='$-d/2<y<d/2$')
plt.plot(y,Ey_yv3(y),'.-',label='$-d/2<y<0$')
plt.xlabel('$y$',fontsize=18)
plt.ylabel('$E_{y}(y)$',fontsize=18)
plt.grid()
plt.xlim(-4,8)
plt.ylim(-10,10)
#plt.ylim(-20,20)
plt.legend(fontsize=15)
plt.title(ur'Funciones por tramos',fontsize=18)
```

```python
def Ey_y(y):
    if y>d/2:
        Ey_y = Ey_y1(y)
    elif -d/2<y<d/2:
        Ey_y = Ey_y2(y)
    elif y<-d/2:
        Ey_y = Ey_y3(y)
    return Ey_y
```

```python
Ey_y(-4),Ey_y(-0.5),Ey_y(0)
```

```python
Ey_yv = np.vectorize(Ey_y)
```

```python
#y = np.linspace(-4,4,1000)
plt.figure(figsize=(6,6))
plt.plot(y,Eyyv(y),'--',lw=3,label=ur'Analítica',color='black')
plt.plot(y,Ey_yv(y),'.-',label='Tramos')
plt.xlabel('$y$',fontsize=18)
plt.ylabel('$E_{y}(y)$',fontsize=18)
plt.grid()
#plt.xlim(-4,8)
plt.ylim(-10,10)
plt.legend(fontsize=15)
plt.title(ur'Expresión definida por tramos',fontsize=18)
```

# Dipolo ideal: Campo eléctrico lejos con $r>>d$

$$\vec{E}(\vec{r})=\frac{k_{e}}{r^5}\left(3(\vec{p}\cdot\vec{r})\vec{r}-r^2 \vec{p}\right)$$

```python
def Elx(x,y):
    return kq*d*(3.*x*y)/(x**2.+y**2.)**(5/2)
def Ely(x,y):
    return kq*d*(3.*y**2-(x**2+y**2))/(x**2.+y**2.)**(5/2)
```

```python
d = 0.1
fig = plt.figure(figsize=(6,6))
Q = plt.quiver(X, Y, Elx(X,Y), Ely(X,Y), units='width')
plt.xlim(-np.pi,np.pi)
plt.ylim(-np.pi,np.pi)
plt.xlabel('$x$',fontsize=18)
plt.ylabel('$y$',fontsize=18)
plt.title(ur'Dipolo eléctrico ideal',fontsize=18)
c1 = plt.Circle((0, d/2.), 0.15, color='red')
c2 = plt.Circle((0, -d/2.), 0.15, color='blue')
plt.gcf().gca().add_artist(c1)
plt.gcf().gca().add_artist(c2)
```

```python
fig = plt.figure(figsize=(6,6))
Q = plt.streamplot(X, Y, Elx(X,Y), Ely(X,Y))
plt.xlim(-np.pi,np.pi)
plt.ylim(-np.pi,np.pi)
plt.xlabel('$x$',fontsize=18)
plt.ylabel('$y$',fontsize=18)
plt.title(ur'Dipolo eléctrico ideal',fontsize=18)
c1 = plt.Circle((0, d/2.), 0.15, color='red')
c2 = plt.Circle((0, -d/2.), 0.15, color='blue')
plt.gcf().gca().add_artist(c1)
plt.gcf().gca().add_artist(c2)
```
