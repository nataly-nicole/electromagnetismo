# Campos Escalares en 2D (Coordenadas Cartesianas)

```python
#Escrito por Nataly Ibarra en marzo de 2017
#Ante cualquier duda o comentario contactar
#a natalynicole.ibarravera@gmail.com
%matplotlib inline
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt
```

En este notebook se han graficado codificando en colores una serie de campos escalares en el plano 2D. 

Se ha elegido para graficar un dominio $x, y\in [-\pi,\pi]$.

```python
x = np.linspace(-np.pi,np.pi)
y = np.linspace(-np.pi,np.pi)
X, Y = np.meshgrid(x,y)
```

# $f_1(x,y)=x$

```python
fig = plt.figure(figsize=(6,6))
plt.pcolormesh(X, Y, X,shading='gouraud')
plt.xlim(-np.pi,np.pi)
plt.ylim(-np.pi,np.pi)
#
plt.xlabel('$x$',fontsize=18)
plt.ylabel('$y$',fontsize=18)
plt.title('$f_1(x,y)=x$',fontsize=18)
#
clb = plt.colorbar()
clb.set_label('$f_1(x,y)$', labelpad=20, y=0.5, rotation=270,fontsize=18)
```

# $f_2(x,y)=y$

```python
plt.figure(figsize=(6,6))
plt.pcolormesh(X, Y, Y,shading='gouraud')
plt.xlim(-np.pi,np.pi)
plt.ylim(-np.pi,np.pi)
#
plt.xlabel('$x$',fontsize=18)
plt.ylabel('$y$',fontsize=18)
plt.title('$f_2(x,y)=y$',fontsize=18)
#
clb = plt.colorbar()
clb.set_label('$f_2(x,y)$', labelpad=20, y=0.5, rotation=270,fontsize=18)
```

# $f_3(x,y)=xy$

```python
plt.figure(figsize=(6,6))
plt.pcolormesh(X, Y, X*Y,shading='gouraud')
plt.xlim(-np.pi,np.pi)
plt.ylim(-np.pi,np.pi)
#
plt.xlabel('$x$',fontsize=18)
plt.ylabel('$y$',fontsize=18)
plt.title('$f_3(x,y)=xy$',fontsize=18)
#
clb = plt.colorbar()
clb.set_label('$f_3(x,y)$', labelpad=20, y=0.5, rotation=270,fontsize=18)
```

# $f_4(x,y)=x^2$

```python
plt.figure(figsize=(6,6))
plt.pcolormesh(X, Y, X**2,shading='gouraud')
plt.xlim(-np.pi,np.pi)
plt.ylim(-np.pi,np.pi)
#
plt.xlabel('$x$',fontsize=18)
plt.ylabel('$y$',fontsize=18)
plt.title('$f_4(x,y)=x^2$',fontsize=18)
#
clb = plt.colorbar()
clb.set_label('$f_4(x,y)$', labelpad=20, y=0.5, rotation=270,fontsize=18)
```

# $f_5(x,y)=y^2$

```python
plt.figure(figsize=(6,6))
plt.pcolormesh(X, Y, Y**2,shading='gouraud')
plt.xlim(-np.pi,np.pi)
plt.ylim(-np.pi,np.pi)
#
plt.xlabel('$x$',fontsize=18)
plt.ylabel('$y$',fontsize=18)
plt.title('$f_5(x,y)=y^2$',fontsize=18)
#
clb = plt.colorbar()
clb.set_label('$f_5(x,y)$', labelpad=20, y=0.5, rotation=270,fontsize=18)
```

# $f_6(x,y)=x^2+y^2$

```python
plt.figure(figsize=(6,6))
plt.pcolormesh(X, Y, (X**2+Y**2),shading='gouraud')
plt.xlim(-np.pi,np.pi)
plt.ylim(-np.pi,np.pi)
#
plt.xlabel('$x$',fontsize=18)
plt.ylabel('$y$',fontsize=18)
plt.title('$f_6(x,y)=x^2+y^2$',fontsize=18)
#
clb = plt.colorbar()
clb.set_label('$f_6(x,y)$', labelpad=20, y=0.5, rotation=270,fontsize=18)
```

# $f_7(x,y)=\sin(x)$

```python
plt.figure(figsize=(6,6))
plt.pcolormesh(X, Y, np.sin(X),shading='gouraud')
plt.xlim(-np.pi,np.pi)
plt.ylim(-np.pi,np.pi)
#
plt.xlabel('$x$',fontsize=18)
plt.ylabel('$y$',fontsize=18)
plt.title('$f_7(x,y)= \sin(x)$',fontsize=18)
#
clb = plt.colorbar()
clb.set_label('$f_7(x,y)$', labelpad=20, y=0.5, rotation=270,fontsize=18)
```

# $f_8(x,y)=\sin(xy)$

```python
plt.figure(figsize=(6,6))
plt.pcolormesh(X, Y, np.sin(X*Y),shading='gouraud')
plt.xlim(-np.pi,np.pi)
plt.ylim(-np.pi,np.pi)
#
plt.xlabel('$x$',fontsize=18)
plt.ylabel('$y$',fontsize=18)
plt.title('$f86(x,y)=\sin(xy)$',fontsize=18)
#
clb = plt.colorbar()
clb.set_label('$f_8(x,y)$', labelpad=20, y=0.5, rotation=270,fontsize=18)
```

# $f_9(x,y)=\sin(x^2+y^2)$

```python
plt.figure(figsize=(6,6))
plt.pcolormesh(X, Y, np.sin(X**2+Y**2),shading='gouraud')
plt.xlim(-np.pi,np.pi)
plt.ylim(-np.pi,np.pi)
#
plt.xlabel('$x$',fontsize=18)
plt.ylabel('$y$',fontsize=18)
plt.title('$f_9(x,y)=\sin(x^2+y^2)$',fontsize=18)
#
clb = plt.colorbar()
clb.set_label('$f_9(x,y)$', labelpad=20, y=0.5, rotation=270,fontsize=18)
```

# Nota: La misma informacin se puede presentar en grficos con superficies como el de a continuacin

# $f_7(x,y)$

```python
Z = np.sin(X)
fig = plt.figure(figsize=(8,8))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap=cm.jet,rstride=3, cstride=3, alpha=1)
#cset = ax.contourf(X, Y, Z, zdir='z', offset=-0.5, cmap=cm.coolwarm)
#cset = ax.contourf(X, Y, Z, zdir='x', offset=-0.2, cmap=cm.coolwarm)
#cset = ax.contourf(X, Y, Z, zdir='y', offset=3.2, cmap=cm.coolwarm)
fig.colorbar(surf, shrink=0.6, aspect=8)#Se agrega barra de colores
ax.set_xlabel('$x$',fontsize=18)
ax.set_ylabel('$y$',fontsize=18)
ax.set_zlabel('$z=f_{7}(x,y)$',fontsize=18)
plt.title('$f_{7}(x,y)=\sin(x)$',fontsize=18)
plt.show()
```
